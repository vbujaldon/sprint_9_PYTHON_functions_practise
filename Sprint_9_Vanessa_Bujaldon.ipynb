#### 1.1 Calculadora IMC (√çndice de masa corporal)
Escribe una funci√≥n que calcule el IMC ingresado por el usuario/a, es decir, quien lo ejecute deber√° ingresar estos datos.
La funci√≥n debe clasificar el resultado en sus respectivas categor√≠as.
Consejo: Intenta validar los datos previamente, para que env√≠e un mensaje de advertencia si los datos introducidos por el usuario/a no est√°n en el formato adecuado o no toma valores razonables.

```python
def pedir_medida(mensaje_medida, minimo=None, maximo=None):
    """
    Solicita a user un valor num√©rico y repite bucle hasta validar que float ok y 
    que est√© dentro del rango de valor m√≠nimo y/o m√°ximo, si se ha indicado.

    Argumentos: 
    - mensaje (str) que se muestra al usuario al solicitar el valor
    - minimo (float o none)
    - maximo (float o none)
    Return: float con valor
    """
    while True:
        try:
            valor = float(input(mensaje_medida))
            if minimo is not None and valor < minimo:
                print(f"Error: el valor debe ser mayor o igual que {minimo}.")
                continue
            if maximo is not None and valor > maximo:
                print(f"Error: el valor debe ser menor o igual que {maximo}.")
                continue
            return valor
            
        except ValueError:
            print(("Error: escribe un valor n√∫merico (y un punto, si deseas decimales). Vuelve a probar."))

def calculo_imc(weight_user, height_user):
    """c√°lculo IMC=peso/(altura**2) con return"""
    return weight_user / (height_user**2)

def mensaje_seg√∫n_imc(imc_calculado):
    """clasificaci√≥n seg√∫n IMC calculado"""
    if imc_calculado > 30:
        mensaje = "Tienes obesidad, contacta a tu m√©dico."
    elif imc_calculado >= 25:
        mensaje = "Tienes sobrepeso, consulta con tu m√©dico."
    elif imc_calculado >= 18.5:
        mensaje = "Tienes un peso normal. Enhorabuena."
    else:
        mensaje = "Tienes un peso demasiado bajo. Contacta a tu m√©dico."
    return mensaje

def main_imc():
    height_user = pedir_medida("Introduce tu altura en metros. Por ejemplo '1.70'.", 0, 3)
    weight_user = pedir_medida("Introduce tu peso en kg. Por ejemplo: '66.5'.", 0, 300)
    imc_calculado = calculo_imc(weight_user, height_user)
    mensaje = mensaje_seg√∫n_imc(imc_calculado)
    print(f'Tu IMC (√≠ndice de masa corporal) es {imc_calculado: .2f}. {mensaje}')

main_imc()
```

#### 1.2. Conversor temperaturas
Selecciona 2 conversores de modo que al introducir una temperatura devuelva, como m√≠nimo, dos conversiones, y de modo que se puedan guardar.
Consejo: Intenta validar los datos previamente, para que env√≠e un mensaje de advertencia si los datos introducidos por el usuario no est√°n en el formato adecuado.

(EXTRA): Piensa una manera de almacenar todas las posibles conversiones en un solo objeto en lugar de escribir muchos if else en funci√≥n de la temperatura de origen y la temperatura de destino.

```python
# funci√≥n validar entrada user (unidades de temperatura concretas)
def pedir_validar_unidad(mensaje, lista_unidades):
    """
    Solicita al usuario la unidad de T¬™ y comprueba que sea v√°lida.
    Repite la petici√≥n hasta que la entrada coincida con una de las unidades permitidas.

    Argumentos: 
        mensaje (string) que muestra al usuario para solicitar la unidad
        lista_unidades (lista) con lista unidades de T¬™ v√°lidas
    Return: string con unidad de medida v√°lida
    """

    while True:
        unidad_origen = input(mensaje).strip().capitalize()
        if unidad_origen in lista_unidades:
            return unidad_origen
        else:   
            print(f"Error: elije la unidad de T¬™ entre las siguientes: {', '.join(lista_unidades)}")

# Funci√≥n conversi√≥n temperatura y guardar resultado en dict
def convertir_temperatura(valor, unidad_origen):
    """
    Convierte una temperatura desde una unidad de origen a las otras unidades disponibles.

    Argumentos: valor (float) y unidad_origen (str)
    Return: (dict) diccionario donde keys: unidades de T¬™ final y values: valores ya convertidos
    """
    
    dic_resultados = {}

    if unidad_origen == "Celsius":
        dic_resultados["Fahrenheit"] = valor * 9/5 + 32
        dic_resultados["Kelvin"] = valor + 273.15

    elif unidad_origen == "Fahrenheit":
        dic_resultados["Celsius"] = (valor - 32) * 5/9
        dic_resultados["Kelvin"] = (valor - 32) * 5/9 + 273.15

    elif unidad_origen == "Kelvin":
        dic_resultados["Celsius"] = valor - 273.15
        dic_resultados["Fahrenheit"] = (valor - 273.15) * 9/5 + 32

    return dic_resultados

def mostrar_resultados(dic_resultado_conversion, temp_usuario, unidad_usuario, repr_unidad_temp):
    print(f"Temperatura original: {temp_usuario}¬∞ {unidad_usuario}, que equivalen a: \n ")
    for unidad_final, valor_convertido in dic_resultado_conversion.items():
        print(f"{unidad_final}: {valor_convertido:.2f} {repr_unidad_temp[unidad_final]}")

def main_conversor_temperaturas():
    # lista para delimitar entrada unidad temperatura
    lista_unidades = ["Celsius", "Kelvin", "Fahrenheit"]
    # diccionario con representaci√≥n corta unidad de Temperatura
    repr_unidad_temp = {
        "Kelvin": "K",
        "Celsius": "¬∫C",
        "Fahrenheit": "¬∫F"
    }
    # llamada funci√≥n validar/input + guardar valores
    temp_usuario = pedir_medida("Introduce la temperatura. Por ejemplo '36.70'.")
    unidad_usuario = pedir_validar_unidad(f"Introduce la medida de origen: {', '.join(lista_unidades)}.", lista_unidades)
    # llamada funci√≥n conversi√≥n unidades y guardar resulta
    dic_resultado_conversion = convertir_temperatura(temp_usuario, unidad_usuario)
    mostrar_resultados(dic_resultado_conversion, temp_usuario, unidad_usuario, repr_unidad_temp)


main_conversor_temperaturas()
```

#### 1.3. Contador de palabras de un texto.

Escribe una funci√≥n que, dado un texto, muestre las veces que aparece cada palabra. Intenta que se gestionen todas las casu√≠sticas que hagan que el programa no funcione correctamente.

(EXTRA): ¬øCu√°l es la longitud media de las palabras del texto que has escrito? ‚Äú¬øHola c√≥mo va?‚Äù deber√≠a devolver (4+3+2) / 3 = 3

#### Versi√≥n SIN COLLECTIONS

```python
def limpieza_simbolos_y_minusculas(texto):
    """ 
    Convierte el texto en min√∫scula y elimina s√≠mbolos de puntuaci√≥n
    Argumento: texto (str) a analizar
    Return: texto (str) ya limpiado
    """
    texto_limpio_minusc = texto.lower() 
    for simbol in ".,;:!?¬ø¬°()[]{}-<>...&*'\"": 
        texto_limpio_minusc = texto_limpio_minusc.replace(simbol, "")
    return texto_limpio_minusc 

def separacion_palabras(texto_limpio_minusc): 
    """ 
    Separa las palabras del texto a partir del espacio entre palabras
    Argumento: texto (str) a procesar
    Return: lista (list) de palabras
    """
    lista_palabras_texto = texto_limpio_minusc.split()
    return lista_palabras_texto 

def recuento_palabras(lista_palabras_texto): 
    """
    Cuantas veces aparece una palabra en la lista
    Argumento: lista de palabras
    Return: diccionario con palabras √∫nicas como claves y frecuencia como valores
    """
    dic_recurrencia_palabras = {} 
    for palabra in lista_palabras_texto: 
        if palabra in dic_recurrencia_palabras: 
            dic_recurrencia_palabras[palabra] += 1 
        else: 
            dic_recurrencia_palabras[palabra] = 1 
    return dic_recurrencia_palabras 

def pedir_texto(mensaje):
    """ 
    Solicita al usuario un texto.
    Argumentos: mensaje (string) que se muestra al usuario al solicitar la entrada.
    Return: string con texto
    """
    return input(mensaje) 

def bucle_print(diccionario, mensaje_print):
    """
    Muestra un texto fijo y recorre el diccionario para mostrar cada par de clave:valor en una l√≠nea.
    Argumento: diccionario, mensaje fijo a imprimir (str)
    Return none (Solo muestra por pantalla)
    """
    print(mensaje_print)
    for clave, valor in diccionario.items():
        print(f"{clave}: {valor}")

mensaje = "Introduce el texto que quieres analizar"
texto = pedir_texto(mensaje)
mensaje_print = "Recuento de palabras en el texto:"

def main_contador_palabras(texto, mensaje_print):
    texto_limpio_minusc = limpieza_simbolos_y_minusculas(texto) 
    lista_palabras_texto = separacion_palabras(texto_limpio_minusc) 
    resultado_recuento = recuento_palabras(lista_palabras_texto) 
    bucle_print(resultado_recuento, mensaje_print)
    return texto

main_contador_palabras(texto, mensaje_print)
```

```python
# EXERCICIO EXTRA: c√°lculo promedio longitud de las palabras

def promedio_long_palabras(lista_palabras_texto):
    """
    Calcula el promedio de letras por palabra.
    Argumento: lista de palabras a analizar (list)
    Return: promedio (float) de letras por palabras, redondeado a 2 decimales
    """
    if len(lista_palabras_texto) == 0: #si lista vac√≠a, evita error division por 0
        return 0 
    total_letras = 0
    for palabra in lista_palabras_texto:
        total_letras += len(palabra)
    return round(total_letras / len(lista_palabras_texto),2)

def main_longitud_palabras(texto):
    # llamada funciones procesado
    texto_limpio_minusc = limpieza_simbolos_y_minusculas(texto) 
    lista_palabras_texto = separacion_palabras(texto_limpio_minusc)
    longitud_texto = promedio_long_palabras(lista_palabras_texto)
    # Muestra resultados:
    print(f"El promedio de la longitud de las palabras del texto: {longitud_texto}")

main_longitud_palabras(texto)
```

#### Versi√≥n usando librer√≠as COLLECTIONS y STRING

```python
from collections import Counter
import string

def limpieza_simb_y_minus_con_string(texto):
    """ 
    Convierte el texto en min√∫scula y elimina s√≠mbolos de puntuaci√≥n
    Argumento: texto (str) a analizar
    Return: texto (str) procesado
    """
    tabla = str.maketrans("", "", string.punctuation)
    return texto.lower().translate(tabla)

def pedir_texto(mensaje):
    """ 
    Solicita al usuario un texto.
    Argumentos: mensaje (string) que se muestra al usuario al solicitar la entrada.
    Return: string con texto
    """
    return input(mensaje) 

def bucle_print(diccionario, mensaje_print):
    """
    Muestra un texto fijo y recorre el diccionario para mostrar cada par de clave:valor en una l√≠nea.
    Argumento: diccionario, mensaje fijo a imprimir (str)
    Return none (Solo muestra por pantalla)
    """
    print(mensaje_print)
    for clave, valor in diccionario.items():
        print(f"{clave}: {valor}")

def main_contador_con_collections(texto_a_analizar, mensaje_print):
    texto_limpio = limpieza_simb_y_minus_con_string(texto_a_analizar)
    palabras = texto_limpio.split()
    resultado = Counter(palabras)
    bucle_print(resultado, mensaje_print)

mensaje_usuario = "Introduce el texto que quieres analizar"
texto = pedir_texto(mensaje_usuario)
mensaje_print = "Recuento de palabras en el texto:"
       

main_contador_con_collections(texto, mensaje_print)
```

#### 1.4. Invertir un diccionario
Intercambia las claves y valores de un diccionario. Los valores y claves en el diccionario original deber√≠an ser √∫nicos pero, en su defecto, la funci√≥n debe imprimir un mensaje de advertencia, junto con una lista con los valores asociados a la clave repetida.

```python
def inversion_diccionario(dic_original):
    """ 
    Intercambio de llaves por valores y viceversa de un diccionario dado. 
    Si encuentra valores duplicados en el diccionario original, 
    muestra un mensaje de alerta y un listado del valor y las llaves donde est√° duplicado.
    
    Argumento: diccionario original
    Return: nuevo diccionario invertido
    """
    
    dic_invertido = {}
    dic_llaves_duplicadas = {}
    for key, value in dic_original.items():
        if value not in dic_invertido:
            dic_invertido[value] = key
        else:   
            if value not in dic_llaves_duplicadas:
                dic_llaves_duplicadas[value] = [dic_invertido[value]]
            dic_llaves_duplicadas[value].append(key)
    if dic_llaves_duplicadas:
        print("ALERTA: hay valores duplicados en el diccionario original que impiden crear todos los registros en el diccionario invertido:")
        for value, key in dic_llaves_duplicadas.items():
            print(f"Valor '{value}' repetido en las siguientes claves: {key}")
    return dic_invertido

def main_inversion_dic(dic_a_analizar):
    dic_final_invertido = inversion_diccionario(dic_a_analizar)
    print("Diccionario invertido:", dic_final_invertido)

diccionario_ejemplo = {"a": "le√≥n", "b":"ciervo", "c": "caballo", "d":"caballo"}
main_inversion_dic(diccionario_ejemplo)
```

```python
# Subir fichero y transformar a string:
def subir_file_pasar_txt_str(link_fichero):
    """
    Abre el fichero en modo lectura (r) y lo convierte a string. Programa el cierre autom√°tico.
    Avisa con mensaje cuando archivo no encontrado.

    Argumento: ruta del fichero.
    Return: fichero en texto (str)
    """
    try:
        with open(link_fichero, "r") as fichero:
            return fichero.read()
    except FileNotFoundError as e:
        print("Error: no se encuentra el archivo", e)
        return ""

# crear dicc dentro de dic con formato {1a letra: {palabras:recuento}}
def agrupar_1a_letra(diccionario):
    """
    Agrupa las palabras por su primera letra, creando un subdiccionario como {letra: {palabra: recuento}}.
    Argumento: diccionario
    Return diccionario con agrupaci√≥n por la primera letra
    """
    dic_agrupado_1a_letra = {}
    for palabra_clave, valor in diccionario.items():
        if not palabra_clave:
            continue # evita IndexError si llave est√° vac√≠a
        primera_letra = palabra_clave[0]
        if primera_letra not in dic_agrupado_1a_letra:
            dic_agrupado_1a_letra[primera_letra] = {}
        dic_agrupado_1a_letra[primera_letra][palabra_clave] = valor
    return dic_agrupado_1a_letra

# Ordenar palabras y transformar a diccionario
def ordenar_palabras_int_ext(diccionario):
    """
    Ordena alfab√©ticamente el diccionario externo (letras iniciales) y los diccionarios internos (palabras).
    Argumento: diccionario.
    Return: diccionario ordenado
    """

    diccionario_ordenado = {}
    for primera_letra, subdic in sorted(diccionario.items()):
        diccionario_ordenado[primera_letra] = dict(sorted(subdic.items())) #dict= crear dic y no lista de tuplas
    return diccionario_ordenado


def main_contar_ordenar_texto(enlace):
    # Ejecutamos subida archivo y conversi√≥n a string
    texto_blanca = subir_file_pasar_txt_str(enlace)
    # Ejecutamos limpieza y recuento palabras (crea dic con "palabra:repeticiones")
    texto_sin_simbolos_minus = limpieza_simbolos_y_minusculas(texto_blanca)
    lista_palabras_separadas = separacion_palabras(texto_sin_simbolos_minus)
    dic_con_recuento = recuento_palabras(lista_palabras_separadas)
    # Ejecutamos crear dic con formato 1 letra + subdic:
    diccionario_agrupado = agrupar_1a_letra(dic_con_recuento)
    # Ejecutamos ordenaci√≥n diccionario
    dic_ordenado = ordenar_palabras_int_ext(diccionario_agrupado)
    # Print en vertical:
    for clave_externa, subdiccionario in dic_ordenado.items(): #items() para mostrar tmb value
        print(f"{clave_externa.upper()}:")
        for clave_interna, valor in subdiccionario.items():
            print(f" {clave_interna}: {valor}")
    return dic_ordenado

main_contar_ordenar_texto(r"C:\Users\vanes\Documents\CURSOS\IT Academy - An√†lisis de dades\_ESPECIALITAT\Sprint 9 Python\tu me quieres blanca.txt")
```

#### 2.2 Conversi√≥n de tipos de datos
El cliente recibe una lista de datos y necesita generar dos listas: la primera, donde estar√°n todos los elementos que pudieron convertirse en flotantes, y la segunda, donde estar√°n los elementos que no pudieron convertirse.

```python
# Transformar lista variada con tuplas y listas a lista simple:
def simplificar_lista_mixta_con_iterables(lista_externa):
    """
    Recorre la lista y devuelve una nueva lista plana con elementos individuales, eliminando los elementos anidados, si los hubiera. 
    
    Argumento: lista_externa (list) que puede contener tuplas, listas o elementos simples.
    Return: (list) nueva lista simplificada. 
    """
    lista_simplificada = []
    for elemento in lista_externa:
        if isinstance(elemento, (list, tuple)):
            lista_simplificada.extend(simplificar_lista_mixta_con_iterables(elemento)) 
            
        else:
            lista_simplificada.append(elemento)
    return lista_simplificada

# Separar elementos seg√∫n si son convertibles a float:
def creacion_listas_float_no_float(lista_externa):
    """
    Clasifica los elementos de una lista seg√∫n si son convertibles a float.

    Argumento: lista_externa que puede contener elementos de distintos tipos.
    Return:
        - lista 1: elementos no convertibles a float.
        - lista 2: elementos convertidos a float.
    """

    elem_float_ok = []
    elem_no_float = []
    for objeto in lista_externa:
        try:
            elem_float_ok.append(float(objeto))
        except (ValueError, TypeError):
            elem_no_float.append(objeto)
    return elem_no_float, elem_float_ok

def main_conversion_datos(lista_externa):
    """ funci√≥n principal que coordina el flujo del programa
    Argumento:
    Return: listas separadas con elementos convertidos a float y los que no
    """
    lista_ready_para_float = simplificar_lista_mixta_con_iterables(lista_externa)
    lista_elem_no_float, lista_float_ok = creacion_listas_float_no_float(lista_ready_para_float)
    print("Lista de elementos convertidos a float:", lista_float_ok)
    print("Lista de elementos NO convertibles a float:", lista_elem_no_float)
    return lista_elem_no_float, lista_float_ok

main_conversion_datos(['1.3', 'one' , '1e10' , 'seven', '3-1/2', ('2',1,1.4,'not-a-number'), [1,2,'3','3.4']])
```

```python
import numpy as np
import string
import pyperclip

# Funci√≥n crear contrase√±a
def crear_password(longitud, mayusc, minusc, numeros, signos):
    """
    Genera contrase√±as aleatorias con los criterios de caracteres especificados como argumentos. 
    
    Argumentos: 
    - longitud (int): N√∫mero de caracteres de la contrase√±a.
    - mayusc (bool): Incluye letras may√∫sculas si es True.
    - minusc (bool): Incluye letras min√∫sculas si es True.
    - numeros (bool): Incluye del 0-9 si es True.
    - signos (bool): Incluye s√≠mbolos especiales si es True.
    Return: (str) cadena de car√°cteres con la contrase√±a
    """

    lista_caracteres = "" # crea cadena vac√≠a
    
    if mayusc: # comprueba si True (hay)
        lista_caracteres += string.ascii_uppercase
    if minusc:
        lista_caracteres += string.ascii_lowercase
    if numeros:
        lista_caracteres += string.digits
    if signos:
        lista_caracteres += string.punctuation
    if not lista_caracteres:
        raise ValueError("Selecciona como m√≠nimo un tipo de caracteres para generar la contrase√±a")
    
    password = ''.join(np.random.choice(list(lista_caracteres), size=longitud))
    return password

def main_crear_contrase√±a():
    # Ejecutamos crear contrase√±a
    password_generado = crear_password(10, True, True, True, True)
    print(password_generado)
    # EXTRA- funci√≥n para copiar autom√°ticamente el texto
    pyperclip.copy(password_generado)

main_crear_contrase√±a()
```

#### 3.2 Procesado de datos simple

Fichero con hist√≥rico de partidos de f√∫tbol, con nombres de equipos y resultados. Necesita que procesemos los datos de forma autom√°tica, para extraer los resultados que necesita. Utiliza el archivo "historic_partits.txt"

El programa debe devolver:

- El n√∫mero total de goles que ha marcado cada equipo.
- El nombre del equipo m√°s goleador.
- El nombre del equipo m√°s goleado.
- La clasificaci√≥n global (cada victoria: 3 pts, empate 1 pts, derrota 0 pts).

```python
import re 

# Abrir fichero y transformar a lista de strings:
def abrir_pasar_file_lista_lineas(link_carpeta):
    """
    Abre el fichero en modo lectura (r) y lo convierte a una lista de l√≠neas. 
    Programa el cierre autom√°tico.
    Avisa con mensaje cuando no se encuentra el archivo.

    Argumento: ruta del fichero.
    Return: lista de l√≠neas (list)
    """
    try:
        with open(link_carpeta, "r") as fichero:
            return fichero.readlines()
    except FileNotFoundError as e:
        print("Error: no se encuentra el archivo", e)
        return []

# Diccionarios para guardar info
def crear_diccionarios():
    """ 
    Crea los diccionarios vac√≠os necesarios para acumular las estad√≠sticas.
    Return: tres diccionarios vac√≠os (dict)
    """
    dic_goles_hechos = {}
    dic_goles_recibidos = {}
    dic_puntos = {}
    return dic_goles_hechos, dic_goles_recibidos, dic_puntos

def procesar_partido(linea, dic_goles_hechos, dic_goles_recibidos, dic_puntos):
    """
    Procesa la l√≠nea del fichero que representa un partido. 
    Ignora √±a l√≠nea si est√° vac√≠a o tiene m√°s palabras que los datos esperados (3).
    Extrae los nombres de equipo visitante y local, los goles visitante y local, y 
    calcula los puntos de cada equipo seg√∫n si es victoria, derrota o empate.
    Con estos datos actualiza los diccionarios de entrada.

    Argumentos:
    - linea (str): l√≠nea del archivo que contiene la informaci√≥n de un partido.
    - dic_goles_hechos (dict): diccionario con los goles marcados por cada equipo.
    - dic_goles_recibidos (dict): diccionario con los goles recibidos por cada equipo.
    - dic_puntos (dict): diccionario con los dic_puntos obtenidos por cada equipo.

    Return: sin return, actualiza los diccionarios de entrada.
    """

    linea = linea.strip()
    if linea == "":
        return

    datos_diferentes_de_un_partido = [x.strip() for x in re.split("\t+", linea)]
    if len(datos_diferentes_de_un_partido) != 3:
        return

    nom_equipo_local, resultado_goles, nom_equipo_visitante = datos_diferentes_de_un_partido
    try:
        goles_local, goles_visitante = map(int, resultado_goles.split("-"))
    except ValueError:
        return

    # Actualizar diccionario con goles marcados
    dic_goles_hechos[nom_equipo_local] = dic_goles_hechos.get(nom_equipo_local, 0) + goles_local
    dic_goles_hechos[nom_equipo_visitante] = dic_goles_hechos.get(nom_equipo_visitante, 0) + goles_visitante

    # Actualizar diccionario con goles recibidos
    dic_goles_recibidos[nom_equipo_local] = dic_goles_recibidos.get(nom_equipo_local, 0) + goles_visitante
    dic_goles_recibidos[nom_equipo_visitante] = dic_goles_recibidos.get(nom_equipo_visitante, 0) + goles_local

    # A√±adir puntos seg√∫n clasificaci√≥n
    if goles_local > goles_visitante:
        dic_puntos[nom_equipo_local] = dic_puntos.get(nom_equipo_local, 0) + 3
    elif goles_local < goles_visitante:
        dic_puntos[nom_equipo_visitante] = dic_puntos.get(nom_equipo_visitante, 0) + 3
    else:
        dic_puntos[nom_equipo_local] = dic_puntos.get(nom_equipo_local, 0) + 1
        dic_puntos[nom_equipo_visitante] = dic_puntos.get(nom_equipo_visitante, 0) + 1


# C√°lculo de las m√©tricas solicitadas
def calcular_metricas_resultados(dic_goles_hechos, dic_goles_recibidos, dic_puntos):
    """
    Calcula las m√©tricas globales a partir de los diccionarios con estad√≠sticas.

    Argumentos:
    - dic_goles_hechos (dict): goles marcados por cada equipo.
    - dic_goles_recibidos (dict): goles recibidos por cada equipo.
    - dic_puntos (dict): puntos acumulados por cada equipo.

    Return:
    - equipo_mas_goleador (str)
    - equipo_mas_goleado (str)
    - clasificacion (list de tuplas) ordenada descendentemente
    """
    equipo_mas_goleador = max(dic_goles_hechos, key=dic_goles_hechos.get)
    equipo_mas_goleado = max(dic_goles_recibidos, key=dic_goles_recibidos.get)
    clasificacion = sorted(dic_puntos.items(), key=lambda x: x[1], reverse=True)
    return equipo_mas_goleador, equipo_mas_goleado, clasificacion

# Muestra resultados
def muestra_resultados(dic_goles_hechos, dic_goles_recibidos, equipo_mas_goleador, equipo_mas_goleado, clasificacion):
    """
    Muestra por pantalla los resultados finales del an√°lisis de partidos, incluyendo:
    - Goles hechos por equipo en orden descendente
    - Goles recibidos por equipo en orden descendente
    - Equipo m√°s goleador y m√°s goleado y sus valores
    - Clasificaci√≥n global seg√∫n puntos.

    Argumentos:
    - dic_goles_hechos (dict): goles marcados por cada equipo.
    - dic_goles_recibidos (dict): goles recibidos por cada equipo.
    - equipo_mas_goleador (str): nombre del equipo m√°s goleador.
    - equipo_mas_goleado (str): nombre del equipo m√°s goleado.
    - clasificacion (lista de tuplas): clasificaci√≥n ordenada por puntos.
    Return: None (solo imprime)
    """

    print("Goles hechos por equipo:")
    for equipo, goles in sorted(dic_goles_hechos.items(), key= lambda x: x[1], reverse = True):
        print(f"{equipo}: {goles}")

    print("\nGoles recibidos por equipo:")
    for equipo, goles in sorted(dic_goles_recibidos.items(), key = lambda x: x[1], reverse = True):
        print(f"{equipo}: {goles}")

    print(f"\nEquipo m√°s goleador: {equipo_mas_goleador} ({dic_goles_hechos[equipo_mas_goleador]} goles hechos)")
    print(f"Equipo m√°s goleado: {equipo_mas_goleado} ({dic_goles_recibidos[equipo_mas_goleado]} goles recibidos)")

    print("\nClasificaci√≥n global (puntos):")
    for equipo, puntos in clasificacion:
        print(f"{equipo}: {puntos}")

def main_partidos(enlace_archivo):
    """
    Funci√≥n principal del programa:
    - Lee el archivo de partidos.
    - Inicializa los diccionarios de estad√≠sticas.
    - Procesa cada partido del fichero.
    - Calcula las estad√≠sticas.
    - Muestra los resultados por pantalla.
    - Retorna las estad√≠sticas
    Argumento: enalce a archivo que hay que procesar
    Return:
    - dic_goles_hechos (dict): goles marcados por cada equipo.
    - dic_goles_recibidos (dict): goles recibidos por cada equipo.
    - equipo_mas_goleador (str): nombre del equipo m√°s goleador.
    - equipo_mas_goleado (str): nombre del equipo m√°s goleado.
    - clasificacion (lista de tuplas): clasificaci√≥n ordenada por puntos.
    """
    lineas_partido = abrir_pasar_file_lista_lineas(enlace_archivo)
    dic_goles_hechos, dic_goles_recibidos, dic_puntos = crear_diccionarios()
    for linea in lineas_partido:
        procesar_partido(linea, dic_goles_hechos, dic_goles_recibidos, dic_puntos)
    equipo_mas_goleador, equipo_mas_goleado, clasificacion = calcular_metricas_resultados(dic_goles_hechos, dic_goles_recibidos, dic_puntos)
    muestra_resultados(dic_goles_hechos, dic_goles_recibidos, equipo_mas_goleador, equipo_mas_goleado, clasificacion)
    return dic_goles_recibidos, dic_goles_hechos, clasificacion, equipo_mas_goleado, equipo_mas_goleador

main_partidos(r"C:\Users\vanes\OneDrive\Documenten\CURSOS\IT Academy - An√†lisis de dades\_ESPECIALITAT\Sprint 9 Python\historic partits.txt")
```
